```python
mf = df.merge(ef, on="customer_id") \
       .merge(gf, on="application_id") \
       .merge(hf, on="customer_id")
```

```python
sns.countplot(x="status", data=mf)
plt.title("Loan Approval vs Rejection")
plt.show()
```

```python
sns.countplot(x="loan_type", data=mf)
plt.xticks(rotation=45)
plt.title("Loan Types Distribution")
plt.show()
```

```python
sns.countplot(x="loan_type", hue="status", data=mf)
plt.xticks(rotation=45)
plt.title("Loan Type vs Approval Status")
plt.show()
```

```python
sns.countplot(x="gender", hue="status", data=mf)
plt.title("Gender vs Loan Approval")
plt.show()
```

```python
sns.histplot(df["income"], kde=True)
plt.title("Income Distribution")
plt.show()
```

```python
sns.boxplot(x="status", y="credit_score", data=mf)
plt.title("Credit Score vs Loan Approval")
plt.show()
```

```python
sns.scatterplot(x="age", y="loan_amount", data=mf)
plt.title("Age vs Loan Amount")
plt.show()
```

```python
sns.scatterplot(x="income", y="loan_amount", hue="status", data=mf)
plt.title("Income vs Loan Amount")
plt.show()
```

```python
sns.countplot(x="source_channel", hue="status", data=mf)
plt.xticks(rotation=45)
plt.title("Marketing Channel vs Loan Status")
plt.show()
```

```python
mf["source_channel"].value_counts().plot.pie(autopct="%1.1f%%")
plt.title("Marketing Channel Share")
plt.ylabel("")
plt.show()
```

```python
corr = mf.select_dtypes(include="number").corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
